```python
'''Web Scaping Tatoli News Agency by Elisabeth'''

#Import all the required libraries
from requests import get
from bs4 import BeautifulSoup
import csv

#Create a csv file to write on
def writeFile(data):
		with open('TatoliNews_.csv', "a", newline='\n', encoding='utf-8') as file:
			writer = csv.writer(file)
			writer.writerow(data)

#Processing 
def extract_links(url):

	# print(url)
	date, news_headline = '',''

	resp = get(url)
	soup = BeautifulSoup(resp.text, "lxml")


	req_div = soup.findAll('div', class_ = 'td_module_1 td_module_wrap td-animation-stack td-meta-info-hide')

	for div in req_div:

		year = div.find('a').get('href').split('/')[3]
		month = div.find('a').get('href').split('/')[4]
		day = div.find('a').get('href').split('/')[5]

		complete_date = year + month + day

		heading = div.find('a').get('href').split('/')[6].replace('-', ' ')

		details = []

		details.append(complete_date)
		details.append(heading)
		details.append(div.find('a').get('href'))

		print(complete_date,'	', heading, '	', div.find('a').get('href'))
		writeFile(list(map(str, details)))

		print('-' * 100)

	

extract_links('http://www.tatoli.tl/category/covid-19/')

base_url = 'http://www.tatoli.tl/category/covid-19/'

#Pull out all the page numbers containing COVID19 News category
for i in range(2,109):
	extract_links(base_url + '/page/' + str(i))
```
